```python
#--------------UNPACK THE VECTOR STORES------------------------
import os
import zipfile
import shutil
#(A)-----------WORLD BANK VECTOR STORE------------------------
zip_path = "/content/vectorstores_backup.zip"
extract_to = "/content/vectorstores"

# Make sure the destination folder exists
os.makedirs(extract_to, exist_ok=True)

# Unzip
with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall(extract_to)

print("✅ Unzipped successfully into /content/vectorstores")
#(B)-----------WORLD BANK VECTOR STORE------------------------
zip_path = "/content/imf_excel_vectorstore.zip"
extract_to = "/content/imf_excel_vectorstore"

# Make sure the destination folder exists
os.makedirs(extract_to, exist_ok=True)

# Unzip
with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall(extract_to)

print("✅ Unzipped successfully into /content/imf_excel_vectorstore")
```
